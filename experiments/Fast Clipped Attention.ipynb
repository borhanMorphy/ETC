```python
from dataclasses import dataclass
import time

import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F
%matplotlib inline
```

```python
@dataclass
class Config:
    d_in: int
    d_out: int
    r: int = 8 # local attention radius
    k: int = 4 # span of relative positions
```

```python
config = Config(d_in=4, d_out=2, r=8, k=4)
config
```

```python
W_Q = torch.randn(config.d_in, config.d_out)
W_K = torch.randn(config.d_in, config.d_out)
W_V = torch.randn(config.d_in, config.d_out)
E_rel = torch.randn(2 * config.k + 1, config.d_out)
```

```python
def plain_attention(x):
    # x: seq_len x d_in

    Q = x @ W_Q
    K = x @ W_K
    V = x @ W_V

    e = Q @ K.T
    e /= (config.d_out ** 0.5)
    # e: S x S

    z = torch.exp(e)
    z /= z.sum(dim=1, keepdims=True)
    return z
```

```python
def naive_attention(x):
    # x: seq_len x d_in
    seq_len = x.shape[0]
    e = torch.empty((seq_len, seq_len), dtype=torch.float32)
    Q = x @ W_Q
    K = x @ W_K
    V = x @ W_V

    for i in range(seq_len):
        for j in range(seq_len):
            r = (j - i) + config.k
            r = min(max(r, 0), config.k * 2)
            e[i, j] = Q[[i], :] @ (K[[j], :] + E_rel[r, :]).T
    e /= (config.d_out ** 0.5)
    # e: S x S
    return e

    z = torch.exp(e)
    z /= z.sum(dim=1, keepdims=True)
    return z
```

```python
def fast_attention(x):
    # x: seq_len x d_in
    seq_len = x.shape[0]
    window = 2 * config.r + 1

    Q = x @ W_Q
    K = x @ W_K
    V = x @ W_V

    K_padded = F.pad(K, (0, 0, config.r, config.r), "constant", value=0)
    # -k, ..., 0, ..., +k
    rel_pos_ids = torch.arange(-config.r, config.r + 1).clip(
        min=-config.k,
        max=+config.k,
    ) + config.k
    rel_pos_embeds = E_rel[rel_pos_ids]

    ids = torch.arange(window).repeat(
        seq_len,
    ).reshape(
        seq_len,
        window,
    )

    ids += torch.arange(seq_len).reshape(seq_len, 1)
    # ids: seq_len x window

    K_clipped = K_padded[ids]
    # K_clipped: seq_len x window x d_out

    e = torch.bmm(
        Q.view(seq_len, 1, config.d_out),
        (K_clipped + rel_pos_embeds).permute(0, 2, 1),
    ).view(seq_len, window)
    e /= (config.d_out ** 0.5)
    # e: seq_len x window
    return e

    z = torch.exp(e)
    z /= z.sum(dim=1, keepdim=True)
    return z
```

```python
x = torch.randn(100, config.d_in)
```

```python
e_naive = naive_attention(x)
plt.imshow(e_naive);
```

```python
e_fast = fast_attention(x)
plt.imshow(e_fast);
```

## Sanity Check

```python
e_naive_padded = F.pad(e_naive, (config.r, config.r, 0, 0), "constant", value=0)

window = 2 * config.r + 1

seq_len = x.shape[0]

ids_j = torch.arange(window).repeat(
    seq_len
).reshape(
    seq_len, window,
)

ids_j += torch.arange(seq_len).reshape(seq_len, 1)
ids_j = ids_j.flatten()

ids_i = torch.arange(seq_len).repeat_interleave(window)

e_naive_clipped = e_naive_padded[ids_i, ids_j].reshape(seq_len, window)

plt.imshow(e_naive_clipped == e_fast);
```

```python
def run_speed_test(attention_fn, num_tries: int = 5, min_seq_len: int = 10, max_seq_len: int = 100):
    experiments_results = list()
    for seq_len in torch.arange(min_seq_len, max_seq_len):
        elapsed_times = list()
        for _ in range(num_tries):
            x = torch.randn(seq_len, config.d_in)
            s = time.time()
            attention_fn(x)
            elapsed_times.append(time.time() - s)
        avg_elapsed_time = (sum(elapsed_times) / num_tries) * 1000
        experiments_results.append((seq_len, avg_elapsed_time))

    seq_lenths, elapsed_times = list(zip(*experiments_results))
    _, ax = plt.subplots()

    ax.set_ylim([0.0, 100])

    ax.plot(seq_lenths, elapsed_times);

    ax.set(xlabel='Sequence Lenght (S)', ylabel='Time (ms)', title='Time Complexity Graph')
    ax.grid()
```

## Plain Attention Speed

```python
run_speed_test(plain_attention, min_seq_len=10, max_seq_len=100)
```

## Naive Attention Speed

```python
run_speed_test(naive_attention, num_tries=1, min_seq_len=10, max_seq_len=100)
```

## Fast Attention Speed

```python
run_speed_test(fast_attention, min_seq_len=100, max_seq_len=1000)
```
